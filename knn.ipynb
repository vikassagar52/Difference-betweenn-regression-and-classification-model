```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import plotting
from sklearn import datasets
from sklearn.model_selection import KFold,train_test_split
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LinearRegression ,LogisticRegression
from sklearn.svm import SVC
```

```python
iris=datasets.load_iris()
x=pd.DataFrame(iris.data)

y=iris.target
plt.figure(figsize=(10,10))
plotting.scatter_matrix(x)
plt.show()
x=np.array(iris.data)
```

```python

knn=KNeighborsClassifier(n_neighbors=3)
kf=KFold(n_splits=3)
for p,q in kf.split(x):
    x_train,x_test,y_train,y_test=x[p],x[q],y[p],y[q]
    knn.fit(x_train,y_train)
    y_pre=knn.predict(x_test)
    
    
    
```

```python

x=(x[:,3]).reshape(-1,1)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2)
K_nn=KNeighborsClassifier(n_neighbors=3)
K_nn.fit(x_train,y_train)
y_pre=K_nn.predict(x_test)
plt.scatter(x_train,y_train)
plt.scatter(x_test,y_pre,color="g",marker="^")
plt.show()
```

```python
l_nn=LinearRegression()
l_nn.fit(x_train,y_train)
y_pre=l_nn.predict(x_test)
plt.scatter(x_train,y_train)
plt.plot(x_test,y_pre,color="g",marker="^")
plt.show()
```

```python
s_nn=SVC()
s_nn.fit(x_train,y_train)
y_pre=s_nn.predict(x_test)
plt.scatter(x_train,y_train)
plt.scatter(x_test,y_pre,color="g",marker="^")
plt.show()
```
